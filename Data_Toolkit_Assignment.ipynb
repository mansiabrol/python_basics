1. What is NumPy, and why is it widely used in Python

- What it is: NumPy (Numerical Python) is a fundamental library for scientific computing in Python.

-  It provides a powerful N-dimensional array object (ndarray), which is a homogeneous collection of items of the same type, along with functions for operating on these arrays.


Why it's widely used:

- Efficiency: NumPy arrays are much more memory-efficient and faster than Python lists for numerical operations. This is because NumPy arrays store data in contiguous memory locations and use optimized C code for calculations.

- Powerful Array Operations: NumPy offers a vast collection of functions for mathematical, logical, statistical, Fourier transform, and linear algebra operations on arrays.

- Foundation for Other Libraries: Many other scientific computing libraries in Python, like Pandas, SciPy, and scikit-learn, are built on top of NumPy, making it an essential building block.

- Broadcasting: NumPy's broadcasting mechanism makes it easy to perform operations on arrays with different shapes

2. How does broadcasting work in NumPy?

- Broadcasting: NumPy's broadcasting is a way of performing arithmetic operations on arrays with different shapes. The smaller array is "stretched" or "broadcast" to match the shape of the larger array before the element-wise operation is performed.

Rules for Broadcasting:

- Dimension Matching: The dimensions of the arrays are compared from right to left. If the dimensions don't match, they are padded with 1 from the left.

- Size Compatibility: For two dimensions to be compatible, they must either be equal or one of them must be 1.

- Broadcast Application: If one dimension has a size of 1, then it will be stretched to match the size of the other array's dimension during calculations.

Example: Adding a scalar value to a NumPy array. The scalar is "broadcast" to the same shape as the array.

8. What does the term "vectorized operation" mean in NumPy?

- Definition: A vectorized operation is a process where an operation is performed on an entire array at once, rather than looping through each element individually.

How it Works in NumPy: NumPy's operations are implemented using optimized C or Fortran code that can execute directly on the array data without the overhead of Python loops.

Benefits:

Speed: Vectorized operations are much faster than Python loops for numerical calculations.

Code Simplicity: Writing code using vectorized operations is often more concise and easier to read than code with explicit loops.

17. Explain the difference between apply() and map() in Pandas.

- map() (Series-specific):

1.Works on a single Pandas Series.

2.Accepts a function that is applied element-wise to each value in the Series.

3.The function you pass must take a single value as an argument and return a single transformed value.

4.Generally used for element-wise transformations or value mapping.

-apply() (Series or DataFrame):

1.Can be used on a Pandas Series or a DataFrame.

2.When used with a Series, it applies a function element-wise, similar to map().

3.When used with a DataFrame, it applies a function along a specified axis (row-wise or column-wise).

4.The function you pass can take a Series or even a DataFrame (when applied row or column-wise) as input and return a single value or a Series.

5.More flexible than map(), allowing for complex row-wise or column-wise operations, aggregations, and calculations.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

- Memory Efficiency: NumPy arrays store data in contiguous memory blocks, which allows them to quickly access a subset of the array by adjusting memory pointers.

-Vectorized Operations: NumPy slice operations typically return a view (rather than a copy) of the underlying data, thus, avoiding costly copying.

- Native Code: NumPy operations are implemented in optimized C or Fortran code, which is much faster than Python code.

```python
# 4. Calculate the correlation matrix using Seaborn and visualize it with a heatmap

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'A': np.random.rand(100),
        'B': np.random.rand(100),
        'C': np.random.rand(100)}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#7. Program to perform element-wise multiplication of two NumPy arrays

import numpy as np

# Create two NumPy arrays
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

# Perform element-wise multiplication
result = array1 * array2

print("Array 1:\n", array1)
print("Array 2:\n", array2)
print("Element-wise multiplication:\n", result)
```

```python
#8. Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# Generate x data
x = np.linspace(0, 10, 100)

# Generate y data for multiple lines
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

# Create the line plot
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) + cos(x)')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot Example")
plt.legend()
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

# Create a sample DataFrame
data = {'Values': [10, 25, 5, 30, 15, 35, 20]}
df = pd.DataFrame(data)

# Define a threshold
threshold = 20

# Filter rows where 'Values' is greater than the threshold
filtered_df = df[df['Values'] > threshold]

print("Original DataFrame:\n", df)
print("Filtered DataFrame (Values >", threshold, "):\n", filtered_df)
```
